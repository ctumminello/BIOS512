```r
# Load the `palmerpenguins` dataset.

library("palmerpenguins")
```

```r
# What's in the penguins table?

penguins %>% head
```

```r
options(repr.plot.width=5, repr.plot.height=3)

p1= ggplot (penguins, aes(x= body_mass_g, y= flipper_length_mm, color= species))

p1= p1 + geom_point ()

p1
```

```r
options(repr.plot.width=5, repr.plot.height=3)

p2= ggplot (penguins, aes(x= body_mass_g, y= flipper_length_mm, color= island))

p2= p2 + geom_point ()

p2= p2 + scale_color_brewer(type = "qual", palette = "Set2")

p2
```

```r
options(repr.plot.width=9, repr.plot.height=3.5)

p1 + p2 + plot_layout(guides = "collect")
```

```r
options(repr.plot.width=6, repr.plot.height=8)

p= ggplot (penguins, aes(x= body_mass_g, y= flipper_length_mm))

p= p + facet_grid(island~species)

p= p + geom_point ()

p
```

```r
options(repr.plot.width=5, repr.plot.height=4)

p = ggplot(penguins, aes(y= island, fill = species))

p = p + geom_bar(position= "fill")

p= p + scale_fill_brewer(type = "qual", palette = "Set2")

p
```
